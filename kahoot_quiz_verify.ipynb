## Section 5: Function & Scope

```python
# Q29 — Expected: A) 5
x = 5
def f():
    x = 10
f()
print(x)
```

```python
# Q30 — Expected: B) 10
x = 5
def f():
    global x
    x = 10
f()
print(x)
```

```python
# Q31 — Expected: B) [1, 2] [1, 2]
def f(a, b=[]):
    b.append(a)
    return b

x = f(1)
y = f(2)
print(x, y)
```

```python
# Q32 — Expected: B) [1, 2, 3, 4]
def modify(lst):
    lst += [4]

a = [1, 2, 3]
modify(a)
print(a)
```

```python
# Q33 — Expected: A) [1, 2, 3]
def modify(lst):
    lst = lst + [4]

a = [1, 2, 3]
modify(a)
print(a)
```
